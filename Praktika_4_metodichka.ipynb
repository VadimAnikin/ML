```python
import numpy as np
import pandas as pd
```

```python
orders = pd.read_csv('/content/orders.csv - orders.csv')
orders
```

```python
customers = pd.read_csv('/content/customers.csv - customers.csv', index_col ='id')
customers
```

```python
df = pd.merge(orders, customers, how='inner', left_on='customer_id', right_index=True)
df.head()
```

1) Сколько заказов, отправлено первым классом за последние 5 лет?

```python
len(df.query('ship_mode == "First"'))
```

2) Сколько в базе клиентов из Калифорнии?

```python
len(df.query('state == "California"').customer_id.unique())
```

3) Сколько заказов они сделали?

```python
len(df.query('state == "California"'))
```

4) Постройте сводную таблицу средних чеков по всем штатам за каждый год.

```python
pd.pivot_table(df, values = 'sales', index = 'state', columns = 'years', aggfunc=np.mean)
```

```python
data = pd.read_csv('/content/продажи.csv', sep = ';')
data.head()
```

```python
def top(data, n):
  return pd.pivot_table(data, values = ['Quantity', 'Amount'],
                 index = 'Item', 
                 aggfunc = np.sum).sort_values(by = 'Amount', ascending = False).head(n)
```

```python
top(data, 5)
```
